```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
data =pd.read_csv("social_media data.csv")
print(data)
```

```python
sns.histplot(data=data, x='Age', bins=30, kde=True)
plt.title('Distribution of Likes')
plt.show()
```

```python
sns.displot(data["Country"],  kde=True, rug=True, color='r')

plt.show()
```

```python
sns.scatterplot(x="Age", y="Academic_Level", data=data, hue="Gender", style="Gender", size="Gender",palette="turbo",alpha=.7)
plt.show()
```

```python
# Histogram of social media time with KDE
sns.histplot(data['social_media_time_min'], kde=True, bins=30, color='skyblue')
plt.title("Histogram of Social Media Time (in minutes)")
plt.xlabel("Social Media Time (min)")
plt.ylabel("Frequency")
plt.show()
```

```python
# KDE Plot of focus score
sns.kdeplot(data['focus_score'], shade=True, color='orange')
plt.title("KDE Plot of Focus Score")
plt.xlabel("Focus Score")
plt.ylabel("Density")
plt.show()
```

```python
# ECDF Plot of mood score
sns.ecdfplot(data['mood_score'])
plt.title("ECDF Plot of Mood Score")
plt.xlabel("Mood Score")
plt.ylabel("Proportion  x")
plt.show()
```

```python
# Displot of anxiety levels with KDE
sns.displot(data['anxiety_level'], kde=True, bins=25, height=4, aspect=1.5, color='mediumseagreen')
plt.title("Displot of Anxiety Level")
plt.xlabel("Anxiety Level")
plt.ylabel("Density")
plt.show()
```

```python
# Barplot: Average Daily Usage by Gender
sns.barplot(data=data, x='Gender', y='Avg_Daily_Usage_Hours', estimator='mean', palette='pastel')
plt.title("Average Daily Social Media Usage by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Daily Usage (Hours)")
plt.show()
```

```python
# Countplot: Most Used Platform
sns.countplot(data=data, x='Most_Used_Platform', palette='viridis')
plt.title("Most Used Social Media Platform")
plt.xlabel("Platform")
plt.ylabel("Number of Students")
plt.xticks(rotation=45)
plt.show()
```

```python
# Boxplot: Addicted Score by Academic Level
sns.boxplot(data=data, x='Academic_Level', y='Addicted_Score', palette='Set2')
plt.title("Addicted Score by Academic Level")
plt.xlabel("Academic Level")
plt.ylabel("Addicted Score")
plt.xticks(rotation=45)
plt.show()
```

```python
# Violinplot: Digital Wellbeing Score by Gender
sns.violinplot(data=data, x='Gender', y='digital_wellbeing_score', palette='muted')
plt.title("Digital Wellbeing Score by Gender")
plt.xlabel("Gender")
plt.ylabel("Digital Wellbeing Score")
plt.show()
```

```python
# Stripplot: Mood Score by Relationship Status
sns.stripplot(data=data, x='Relationship_Status', y='mood_score', jitter=True, palette='cool')
plt.title("Mood Score by Relationship Status")
plt.xlabel("Relationship Status")
plt.ylabel("Mood Score")
plt.xticks(rotation=15)
plt.show()
```

```python
# Catplot (Box) of Focus Score by Addiction Level
sns.catplot(data=data, x='addiction_level', y='focus_score', kind='box', palette='coolwarm')
plt.title("Focus Score by Addiction Level")
plt.xlabel("Addiction Level")
plt.ylabel("Focus Score")
plt.show()
```
